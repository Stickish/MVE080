### Question 1.1

During Lecture 3 we discussed color palettes for categorical data. Briefly describe and motivate the three key properties we want a good categorical color palette to have.

### Answer 1.1
Color palettes for categorical data should only contain a finite amount of colors, since they should be distinct from each other and equivalent. It is also important that no color stands out relative to the others, since this would give the impression that it is more important than the other colors. In the same vein it is also important that they do not give of the appearance of an order, since there should by definition not exist an order between the categories.

### Question 1.2

In Lecture 3 we used pie- and bar-charts to visualize proportions. Briefly discuss the benefits and drawbacks with using bar-charts and pie-charts to visualize proportions. 

### Answer 1.2

#### Pie-charts
Pie-charts makes it easier to compare counts to the whole compared to bar-charts since they by definition also include the total (it is quite easy to compare the area of one section of the pie-chart to the whole). Pie-charts are especially good at visualizing 'simple' fractions $[\frac{1}{2}, \frac{1}{3}, \frac{1}{4}, \frac{1}{5}]$ etc. One of the drawbacks of a pie-chart is that as the amount of different categories grows, it becomes harder to interpret the chart.



#### Bar-charts
Makes it easy to compare fractions to each other but harder to compare to the whole. Above I mentioned that as the number of categories in a chart grows pie.charts become muddy, this is not the case for bar-charts (within reason).

### Question 2.2 

In Lecture 3 when comparing colormaps (e.g slide 2) I plotted this function;

$$
    z = 0.1 \mathrm{sin}(r) + \mathrm{cos}(r^2) \mathrm{exp}\Big(-\frac{r}{2 \pi} \Big) \ \mathrm{where} \\
    r = x^2 + y^2 \quad x, y \in [-4\pi, 4\pi],
$$

where I used colors to represent $z$. 

Now plot (as in Lecture 3 slide 2) the function

$$
    z = 0.1 \mathrm{sin}(r) + \mathrm{cos}(r^2) \mathrm{exp}\Big(-\frac{x}{4 \pi} \Big) \ \mathrm{where} \\
    r = x^2 + y^2 \quad x, y \in [-4\pi, 4\pi],
$$

using three different [colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html) that are sufficiently distinct. Using three monochrome maps (e.g Greys, Blues and Purples) does not count as sufficiently distinct. Code for how to use the rainbow desaturated colormap is in the Lecture 3 notebook on Canvas.

For each colormap briefly discuss its benefits based on what you can see from the plots you have created.

```python
x = np.linspace(-4*np.pi, 4*np.pi, 200)
y = np.linspace(-4*np.pi, 4*np.pi, 200)
xx, yy = np.meshgrid(x,y)
rr = xx**2 + yy**2
zz = 0.1*np.sin(rr) + np.cos(rr**2)*np.exp(-xx/(4*np.pi))

xdf = xx.flatten()
ydf = yy.flatten()
rdf = rr.flatten()
zdf = zz.flatten()
df = pd.DataFrame({'x':xdf, 'y':ydf, 'r':rdf, 'z':zdf})
```

```python
# Insert code here for colormap 1


p = (ggplot(df, aes('x', 'y', fill='z'))
    + geom_raster()
    + scale_fill_cmap('Oranges')
    + scale_x_continuous(expand=(0, 0)) 
    + scale_y_continuous(expand=(0, 0))
    + theme_minimal()
)
p
# p.save("q22_oranges.png", dpi=300, width=BASE_WIDTH*1.1, height=BASE_WIDTH)
# Remember to print the visual 
```

```python
# Insert code here for colormap 2
p = (ggplot(df, aes('x', 'y', fill='z'))
    + geom_raster()
    + scale_fill_cmap('viridis')
    + scale_x_continuous(expand=(0, 0)) 
    + scale_y_continuous(expand=(0, 0))
    + theme_minimal()
)
p
# p.save("q22_viridis.png", dpi=300, width=BASE_WIDTH*1.1, height=BASE_WIDTH)
# Remember to print the visual 
```

```python
# Insert code here for colormap 3
col_list_df = pd.read_csv("Rainbow_desaturated.csv")
rainbow_desaturated = []
for i in range(col_list_df.shape[0]):
    rainbow_desaturated.append([col_list_df["R"][i], col_list_df["G"][i], col_list_df["B"][i]])

p = (ggplot(df, aes('x', 'y', fill='z'))
    + geom_raster()
    + scale_fill_gradientn(colors=rainbow_desaturated)
    + scale_x_continuous(expand=(0, 0)) 
    + scale_y_continuous(expand=(0, 0))
    + theme_minimal()
)
print(p)
# p.save("q22_rainbow.png", dpi=300, width=BASE_WIDTH*1.1, height=BASE_WIDTH)
# Remember to print the visual 
```

*Oranges*

Using the oranges, or similar, colormap it is easy to see the trend of the values on the right getting closer together as x grows. However it is harder to notice differences between values.

*viridis*

The values getting closer and closer to zero on the right half of the plot is still visible however it is a little bit easier to see the difference between values on the left half compared to the oranges colormap.

*Desaturated rainbow*

The Shift in color between right and left becomes clearer and it is much easier to see the difference between values, especially on the left.

### Question 3.3 

The connected scatter plot (Fig. 13.10 in the [course book](https://clauswilke.com/dataviz/) is an approach that allows us to visualize two separate time-dependent variables. On the webpage I have uploaded an example for a connected scatter plot where I plot unemployment rate and apartment prices for Sweden from 1996-2020. Use the dataset on Canvas to recreate this visual. Briefly comment what you think is the main takeaway message from the plot. 

```python
# Insert code here 
df = pd.read_csv('House_unemployed.csv')
years = [1996, 2001, 2008, 2011, 2018]
breaks = df.query('Year in @years')

p = (ggplot(df, aes(x='price/1000', y='Unemployment', color='Year'))
    
    +geom_path(show_legend=False, size=1)
    +geom_point(show_legend=False, size=2)
    +labs(x='Price [kSEK]', y='Unemployment [%]', title='Sweden : Apartment prices vs unemployment')
    +geom_text(data=breaks,  
               mapping=aes('price/1000', 'Unemployment', label=['1996', '2001', '2008','2011', '2018']),
               show_legend=False,
               nudge_y=-0.4,
               color='black'
              )
     +annotate('text', x=40.9, y=8.7, label='2020')
     +coord_flip()
     +theme_minimal()
    )
print(p)
# Remember to print the visual 
```

### Question 5.2 

Let us explore the dataset in question 5.1 a bit deeper. Choose five different variables and create a visual where we can clearly see the associations between all pair of variables. Briefly discuss the take-away from the visual. In the plot color the different continents.

```python

# Reread the data
df = pd.read_csv('World_2018-1.csv')
# print(df)
plot_vars = ['GDP_per_capita', 'Freedom_to_make_life_choices', 'Generosity', 'Perceptions_of_corruption', 'Social_support']
df1 = df[['name', 'continent'] + plot_vars]
df_tmp1 = df[['name', 'continent'] + plot_vars]
df_tmp = df_tmp1.melt(id_vars=['name', 'continent'], 
                      value_vars=plot_vars, value_name='val_x', var_name='var_x')
df_tmp = df_tmp.merge(df_tmp1)
df_plot = df_tmp.melt(id_vars=["name", "continent", "var_x", "val_x"], 
                      value_vars=plot_vars, value_name="val_y", var_name="var_y")


p = (ggplot(df_plot, aes("val_x", "val_y", color="continent"))
    + geom_point()
    + scale_color_manual(values=cbPalette, name = "Continent")
    + facet_grid("var_y ~ var_x", 
                 scales="free")
    + labs(x = "", y = "")
    + theme_bw()
    + theme(legend_position="bottom", figure_size=(10,10)))
p
# Remember to print the visual 
```

It is hard to be certain about the conclusions drawn from this plot however there seem to be some correlation between social support and gdp per capita, there also exist some correlation between generosity and freedom of choice. 

These findings also makes some sense to us humans since a higher gdp should indicate that there exists more systems for providing social support. That generosity and freedom of choice are somewhat correlated is also no huge surprise. However it is dangerous to walk down this path without caution, since we are prone to find these patterns in data when they might not exist at all due to us being human.

### Question 5.3 

Has the amount of $\mathrm{CO}_2$ released per person increased from 1970 to 2018 for a majority of countries, and if yes what magnitude of increase are we talking about? Use the $\mathrm{CO}_2$ on Canvas and create a visual which answers these questions.

```python
# Insert code here 
df = pd.read_csv("CO2_data.csv")

df["diff"]  = df["CO2_2018"] - df["CO2_1970"]
                                                    
df["color"] = np.sign(df["diff"])
                                                    
df = df.sort_values("diff")

pos = df["Country"].values[np.argsort(df["diff"])]

df = df.assign(color = pd.Categorical(df["color"]))

col = [cbPalette[1], cbPalette[5]]

num_decrease = df[df["color"]== -1].shape[0]
num_increase = df[df["color"]== 1].shape[0]

p = (ggplot(df, aes("Country", "diff", group = 1))
            + geom_point(size = 1)
            + geom_path()
            + geom_area(aes(ymin = 0, where = "color == 1.0"), fill = cbPalette[5], alpha=0.7)
            + geom_area(aes(ymax = 0, where = "color == -1.0"), fill = cbPalette[4], alpha=0.7)
            + geom_hline(yintercept = 0)
            + scale_x_discrete(limits = pos)
            + annotate("text", x= 60, y=-25, size=16, label=f"{num_decrease} countries decreased their carbon dioxide footprint")
            + annotate("text", x= 120, y=15, size=16, label=f"{num_increase} countries increased their carbon dioxide footprint")
            + theme_minimal()
            + labs(x='', y='Change in $\mathrm{CO}_2$ released per person', title='Change in $\mathrm{CO}_2$ released between 1970 and 2018 for 180 countries ')
            + theme(figure_size=(16, 10), axis_text_x = element_blank())
    )

p
# Remember to print the visual 
```
