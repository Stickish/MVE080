### Question 1.3
Describe situations when _nonlinear axes_ might be useful.
When should they not be used?

### Answer 1.3
Nonlinear axes like the logarithmic one is useful when points are spaced very close together, taking the logarithm will space them out more. It is also useful when depicting exponential data, since it makes it easier to see "how" exponential a line is for example. The logarithmix transformation is also useful when plotting ratios since a linear scale obfuscates the differences between two datapoints with a ratio $<1$ for example 0.01 and 0.05 can be hard to tell apart if the scale goes from $[0,1000]$.

Logartithmic scales should be avoided when zeros occur in your data since $\mathrm{ln}(0) \rightarrow\infty$.

### Answer 1.4

### Question 3.2

In Lecture2 I used boxplots, error-bars, and violin plots to visualize fluctuations in winter temperature in Västerås. Now using the weather data above plot the temperature per month (like in Fig. 9.8 [here](https://clauswilke.com/dataviz/boxplots-violins.html)) using error-bars, boxplots and violin plots with data points. For each month plot the temperature for 2009 and 2010 next to each other (see example on webpage). 

```python
# Insert code here using standard error
df = data_tidy
# df = df.assign(year=pd.Categorical(df['year']))
# df = df.assign(month=pd.Categorical(df['month']))

months = df['month'].unique()
df9 = pd.DataFrame()

df9['month'] = months

for m in months:
    m9 = df[df['year']==2009]
    m10 = df[df['year']==2010]
    
    mean_9 = m9.loc[m9['month']==m,'Temp'].mean()
    mean_10 = m10.loc[m10['month']==m,'Temp'].mean()
    
    std_9  = m9.loc[m9['month']==m,'Temp'].std()
    std_10 = m10.loc[m10['month']==m,'Temp'].std()
    
    df9.loc[df9['month']==m,'mean_T'] = mean_9
    df10.loc[df9['month']==m,'mean_T'] = mean_10
    
    df9.loc[df9['month']==m,'std'] = std_9
    df10.loc[df9['month']==m,'std'] = std_10
    
    df9.loc[df9['month']==m,'m-s'] = mean_9 - std_9
    df10.loc[df9['month']==m,'m-s'] = mean_10 - std_10
    
    df9.loc[df9['month']==m,'m+s'] = mean_9 + std_9
    df10.loc[df10['month']==m,'m+s'] = mean_10 + std_10
  
df9['year'] = 2009
df10['year'] = 2010
# print(df9)

df = pd.concat([df9, df10])
  
df = df.assign(month=pd.Categorical(df['month']))
df = df.assign(year=pd.Categorical(df['year']))


g = (ggplot(df,aes("month", "mean_T", group='year', color='year')) 
     + geom_point(position=position_dodge(0.5)) 
     + geom_errorbar(aes(x="month", ymin="mean_T - std",ymax="mean_T + std"), position=position_dodge(0.5))
     + scale_x_discrete(limits = months)
     + ylab('Temperature')
     + xlab('Month')
     + ggtitle('Mean temperatures')
     + theme_minimal()
    )
g
```

Briefly discuss the drawback with each approach (standard errors, boxplots, and violin plots). 

Standard errors:
This approach is good when you are visualizing data that you know is gaussian. Otherwise it can be quite easy to draw faulty conclusions from the plot.

Boxplots:
If the data is distributed in two groups the boxplot will give a faulty description of the distribution, since it will look like the most points are in the middle. 

Violinplots:
Like the types above violinplots with few datapoints can make one draw the wrong conclusions of the distribution.
